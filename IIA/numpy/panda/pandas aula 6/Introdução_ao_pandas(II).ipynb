## 1.&nbsp;vamos começar por importar a biblioteca e o dataset

```python
# importar a biblioteca
import pandas as pd
import numpy as np
```

```python
# importar o dataset que está em csv
import pandas as pd


df_sales = pd.read_csv('supermarket_sales.csv')
print(df_sales)
```

## 3.&nbsp;vamos sumarizar os dados numéricos

```python
# vamos criar um dataframe só com as colunas numéricas
df_select_dtypes = df_sales.select_dtypes(include = ["float64"])
df_select_dtypes.head()
```

```python
# vamos calcular a média de cada uma das colunas
df_select_dtypes.mean() # axis=0 | axis=1 | axis=None
```

```python
df_select_dtypes.mean(axis=None)
```

```python
# vamos calcular a média de cada uma das colunas
np.mean(df_select_dtypes) # axis=0 | axis=1 | axis=None
```

```python
# vamos calcular a média do Unit price
df_select_dtypes["Unit price"].mean() # .min(); .max(); .std()
```

```python
# vamos calcular a média do Unit price
np.mean(df_select_dtypes["Unit price"]) # .min(); .max(); .std()
```

```python
# vamos calcular a média do Unit price usando o método .agg()
df_select_dtypes["Unit price"].agg("mean")
```

```python
# vamos calcular a média do Unit price usando o método .agg()
df_select_dtypes["Unit price"].agg(np.mean)
```

```python
# para criar estatísticas personalizadas -> .agg()
# vamos definir uma função
def my_mean(column):
  return column.sum()/(column.count()-1)

# vamos aplicar a nossa função a uma coluna
df_select_dtypes["Unit price"].agg(my_mean)
```

```python
# vamos calcular a soma do Total
df_select_dtypes["Total"].sum()
```

```python
# vamos calcular a transação mais recente
df_sales["Date"].max()
```

```python
# posso aplicar a mais do que uma coluna
df_select_dtypes[["Unit price", "gross margin percentage"]].agg("mean")
```

```python
# posso aplicar várias funções às mesmas coluna
df_select_dtypes[["Unit price", "gross margin percentage"]].agg(["mean", my_mean])
```

```python
# posso aplicar diferentes funções em diferentes colunas
df_select_dtypes.agg({"Unit price": ["mean", my_mean], "gross margin percentage": "mean"})
#                      colunas      lista de funções      dicionário com colunas e funções
```

## 4.&nbsp;Contar

```python
# vamos eliminar duplicados considerando apenas a coluna branch
df_sales_unique = df_sales.drop_duplicates(subset = ["Branch"])
df_sales_unique
```

```python
# vamos contar quantas linhas temos por "Branch"
df_sales["Branch"].value_counts() # .reset_index()
```

```python
# vamos contar quantas linhas temos por invoice
df_sales["Invoice ID"].value_counts()
```

```python
# vamos ordenar as contagens de linha de produtos
df_sales["Product line"].value_counts(sort = True, ascending = False) # descendente por default
```

```python
# podemos ver este valor por frequencia relativa
df_sales["Product line"].value_counts(normalize = True)
```

## 5.&nbsp;Agrupar

```python
# vamos ver a média de vendas por cidade
df_sales.groupby("City")["Total"].mean() # .reset_index()

# cada cidade e a média de vendas correspondente
```

```python
# vamos ver o máximo de quantidade vendida por linha de produto
df_sales.groupby("Product line")["Quantity"].sum()
```

```python
# podemos ter mais do que uma função e variável
df_agg = df_sales.groupby("Product line")[["Quantity", "Total"]].agg(["min", "max", "std"]).reset_index()
df_agg
```

```python
# df_agg tem os nomes das colunas como multi index
df_agg # .columns
```

```python
df_agg.columns = ["_".join(col).strip("_") for col in df_agg.columns] # list comprehension
df_agg
```

```python
# podemos ter mais do que um elemento em cada uma das opções
df_sales.groupby(["City", "Product line"])[["Quantity", "Total"]].agg(["min", "max"])
```

```python
# podemos ter mais do que um elemento em cada uma das opções
df_sales.groupby(["City", "Product line"]).agg({"Total":["sum", "mean"], "Quantity": "max"})
```

## 6.&nbsp;Tabelas Pivot

```python
df_sales
```

```python
# vamos ver como fazer uma tabela pivô
df_sales.pivot_table(values = "Total", index = "City") # .reset_index()
```

```python
# agora queremos a soma e não a média (q é por default)
df_sales.pivot_table(values = "Total", index = "City", aggfunc = "sum")
```

```python
# também podemos usar funções do numpy
df_sales.pivot_table(values = "Total", index = "City", aggfunc = np.sum)
```

```python
# podemos ter mais do que uma função
df_sales.pivot_table(values = "Total", index = "City", aggfunc = [np.sum, np.mean])
```

```python
# vamos agora ver a informação, mas por cidade e linha de produto
df_sales.pivot_table(values = "Total",
                     index = "City",
                     columns = "Product line",
                     aggfunc = "sum")
```

```python
# vamos agora adicionar os totais
df_sales.pivot_table(values = "Total",
                     index = "City",
                     columns = "Product line",
                     aggfunc = "sum",
                     margins = True).astype(int)

# margin = True -> adiciona totais
```
