```python
# crie um df com a categoria das bebidas (Beverage_category) e as calorias
# tenho que fazer uma lista de lista do índices (categoria e caloria) pra criar o df

df_beverage_cg_and_calories = pd.DataFrame(df_starbucks[["Beverage_category", "Calories"]])
print(df_beverage_cg_and_calories)
```

```python
# filtre as bebidas com calorias acima de 400
condition = df_beverage_cg_and_calories["Calories"] > 400
df_starbucks[condition]
```

```python
# filtre as bebidas com calorias acima de 400
# mas que não sejam "Signature Espresso Drinks"
condition_1 = df_beverage_cg_and_calories["Calories"] > 400
condition_2 = df_beverage_cg_and_calories["Beverage_category"] != "Signature Espresso Drinks"
df_starbucks[condition_1 & condition_2]
```

```python
# ordene as bebidas pela suas calorias (Calories) -> ascendente
df_starbucks.sort_values("Calories")

# ordene as bebidas pela suas calorias (Calories) -> descendente
df_starbucks.sort_values("Calories", ascending= False)

# ordene as bebidas pelas calorias-> desc e pelas proteínas ( Protein (g))-> asc
df_starbucks.sort.values(["Calories", "Protein"], ascending = [False, True])

# crie uma serie com a coluna das calorias (Calories)
calories_series = pd.Series(df_starbucks["Calories"])

# crie um dataframe com a mesma coluna
calories_df = pd.DataFrame(calories_series)

# crie um df com a categoria das bebidas (Beverage_category) e as calorias
df_beverage_cg_and_calories = pd.DataFrame(df_starbucks[["Beverage_category", "Calories"]])

# ordene este novo df pelas categorias-> asc e pelas calorias-> desc
df_beverage_cg_and_calories.sort_values(["Beverage_category", "Calories"], ascending = [True, False])

# filtre as bebidas com calorias acima de 400
condition = df_beverage_cg_and_calories["Calories"] > 400
df_starbucks[condition]

# filtre as bebidas com calorias acima de 400
# mas que não sejam "Signature Espresso Drinks"
condition_1 = df_beverage_cg_and_calories["Calories"] > 400
condition_2 = df_beverage_cg_and_calories["Beverage_category"] != "Signature Espresso Drinks"
df_starbucks[condition_1 & condition_2]
```

```python
# queremos saber -> açucar/carbs*100
df_starbucks["Sugar_Carbs_ratio"] = df_starbucks["Sugars (g)"]/df_starbucks["Total Carbohydrates (g)"]*100

# queremos apenas ficar com as colunas:
#(Beverage, Calories, Sugar_Carbs_ratio, Caffeine (mg))
list_of_columns = ["Beverage", "Calories", "Sugar_Carbs_ratio", "Caffeine (mg)"]
df_starbucks_subset = df_starbucks[list_of_columns]

# ordene pelas calorias -> descendente
df_starbucks_subset_sort = df_starbucks_subset.sort_values("Calories", ascending= False)
df_starbucks_subset_sort
```
